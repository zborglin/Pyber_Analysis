```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, label='Boston')
ax.set_xlabel('x label')
ax.grid()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# set the limit between 0 and 45
plt.ylim(0, 45)

# Create title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
#Axes labesl
plt.xlabel("Date")
plt.ylabel("Fare ($)")
#Title
plt.title("PyBer Fare by Month")
#Legend
plt.legend()
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis, color='magenta')
plt.gca().invert_yaxis()
plt.legend('Boston')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title('PyBer by Month')
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='cyan', label='Chicago')
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
```

```python
plt.plot(y_axis, x_axis, 'o', color='red', label='Chicago')

plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.legend()
plt.title('PyBer Fare by Month')
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis, label='Boston')
ax.grid()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
